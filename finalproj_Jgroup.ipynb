```
df = pd.read_csv('A_stat_data_112.csv')
```

```
  file_path = "A_stat_data_112.csv"
  df = pd.read_csv(file_path)

  # 查看數據基本信息
  print(df.head())
  print(df.info())
```

```
  # 選擇需要的欄位
  needed_columns = ['grade', 'is_completed', 'correct_rate', 'answer_duration_sec']
  df = df[needed_columns]

  # 刪除包含缺失值的行
  df = df.dropna()
```

```
  # 確保數據格式正確
  df['is_completed'] = df['is_completed'].astype(int)
  df['correct_rate'] = df['correct_rate'].astype(float)
  df['answer_duration_sec'] = df['answer_duration_sec'].astype(float)

  # 按年級分組計算指標
  grade_stats = df.groupby('grade').agg(
      completion_rate=('is_completed', 'mean'),
      average_correct_rate=('correct_rate', 'mean'),
      average_answer_time=('answer_duration_sec', 'mean')
  ).reset_index()

  print(grade_stats)
```

```
  from sklearn.cluster import KMeans
  from sklearn.preprocessing import StandardScaler

  # 聚類數據標準化
  features = grade_stats[['completion_rate', 'average_correct_rate', 'average_answer_time']]
  scaler = StandardScaler()
  scaled_features = scaler.fit_transform(features)

  # KMeans 聚類分析
  kmeans = KMeans(n_clusters=3, random_state=42)
  grade_stats['cluster'] = kmeans.fit_predict(scaled_features)

  print(grade_stats)
```

```
  import matplotlib.pyplot as plt
  import seaborn as sns

  # 完成率柱狀圖
  plt.figure(figsize=(8, 6))
  sns.barplot(data=grade_stats, x='grade', y='completion_rate', palette='Blues')
  plt.title('Completion Rate by Grade')
  plt.xlabel('Grade')
  plt.ylabel('Completion Rate')
  plt.show()

  # 正確率柱狀圖
  plt.figure(figsize=(8, 6))
  sns.barplot(data=grade_stats, x='grade', y='average_correct_rate', palette='Greens')
  plt.title('Correct Rate by Grade')
  plt.xlabel('Grade')
  plt.ylabel('Correct Rate')
  plt.show()

  # 答題時間柱狀圖
  plt.figure(figsize=(8, 6))
  sns.barplot(data=grade_stats, x='grade', y='average_answer_time', palette='Oranges')
  plt.title('Answer Time by Grade')
  plt.xlabel('Grade')
  plt.ylabel('Average Answer Time (seconds)')
  plt.show()
```
