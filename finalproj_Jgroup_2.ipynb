```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'age': ['18-24y', '25-34y', '35-44y', '45-54y', '55y_up'],
    'people': [330, 510, 255, 195, 210]
}

df = pd.DataFrame(data)

# 使用 seaborn 繪製柱狀圖
sns.barplot(x='age', y='people', data=df, palette='Blues_d')
plt.title('age&people_range')
plt.show()

```

```
import pandas as pd

# 創建資料
data = {
    'source': ['Paper books', 'e-book', 'audiobook', 'Online articles or novels', 'other'],
    'Proportion(%)': [34, 23, 5, 36, 2]
}

df = pd.DataFrame(data)

# 計算平均值和標準差
mean_value = df['Proportion(%)'].mean()
std_deviation = df['Proportion(%)'].std()

# 輸出結果
print(f"平均值: {mean_value}")
print(f"標準差: {std_deviation}")
```
